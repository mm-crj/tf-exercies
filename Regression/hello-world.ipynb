```python
import tensorflow as tf 
import numpy as np
from tensorflow import keras
```

### Definitition of the neural network

The first line defines the nework. `Dense` in `keras` means a single layer of connected neurons. This is the simplest kind of network, a single layer of connected neurons. Succesive layers are defined in sequence hence the name `Sequential`. `input_shape` defines the shape of the input to this layer, which is just a scalar in this case. The `units` define dimensionality of the output space.  [Docs](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense#arguments)

The next line defines the loss function and the optimizer for the training. In this example the optimizer is `sgd` or [stocastic gradient descent](https://en.wikipedia.org/wiki/Stochastic_gradient_descent) and the loss function is `mean_squared_error` 

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

### Data

```python
xs=np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],dtype=float)
ys=np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0],dtype=float)
```

### Fitting

It will interate the optimizer for `epochs` number of times.

```python
model.fit(xs,ys,epochs=500)
```

### Prediction

We have trained the model on the given training data. Now we try to predict an unknown value 10.0. Notice the prediction is not exactly 19, its very near to it. This may be due to a variety of reasons, such as very less training data and the prediction is a probabilistic one not a deterministic one. 


```python
print(model.predict([10.0]))
```
